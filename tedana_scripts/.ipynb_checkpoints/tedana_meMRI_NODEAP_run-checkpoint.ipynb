```python
%matplotlib inline
import os.path as op
import os
from glob import glob
from tedana import workflows
import numpy as np
import nibabel as nib
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from nilearn import plotting, masking
from nilearn.image import math_img
import time
import fnmatch
```

```python
!tedana --help
```

```python
echo_times = np.array([14.60, 39.04, 63.48])
HomeDir = '/Volumes/X9Pro/NODEAP/MRI';
```

```python
pattern = "NODEAP*"
sub_id_list = [item for item in os.listdir(HomeDir) if fnmatch.fnmatch(item, pattern)]
sub_id_list = [item for item in sub_id_list if os.path.isdir(os.path.join(HomeDir, item))]
sub_id_list = sorted(sub_id_list)
sub_id_list
```

```python
rest_names = ['D0', 'S1D1', 'S1D2', 'S2D1', 'S2D2', 'S3D1', 'S3D2']
```

```python
for SubID in sub_id_list:
    SubDir = os.path.join(HomeDir,SubID);
    for curr_rest in rest_names:
        print("SubID: {}".format(SubID))
        print("curr_rest: {}".format(curr_rest))
        DATA_DIR = os.path.join(SubDir,"nifti",curr_rest + "_me")
        if os.path.exists(DATA_DIR) and os.path.isdir(DATA_DIR): # if nii folder exists
            files = glob(os.path.join(DATA_DIR,"r*.nii")) # find all such files: resliced files
            files = sorted(files);                        # sort files so they are in order of e1,e2,e3
            out_dir = os.path.join(SubDir,"tedana_output",curr_rest,'aic')
            try:
                os.makedirs(out_dir) # parent dir does not need to exist
            except FileExistsError:
                pass
            report_file = os.path.join(out_dir,'tedana_report.html')
            if os.path.exists(report_file) and os.path.isfile(report_file): # if report_file already exists
                print("curr_rest: {} already done".format(curr_rest))
            else:
                workflows.tedana_workflow(
                    files,
                    echo_times,
                    out_dir=out_dir,
                    fittype='loglin',    
                    combmode='t2s', # or 'paid'
                    tedpca='aic', # options: 'aic','mdl','kic','kundu','kundu-stabilize'
                    overwrite=True,
                    verbose=True)
```
